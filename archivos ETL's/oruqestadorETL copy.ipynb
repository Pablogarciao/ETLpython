```python
df_parents = df_codes[df_codes['Level'] == 2].copy() #copia para q no se altere la fuente original a pesar de q ahí no se esté modificando
df_parents
```

Transformación

```python
#1. Limpiar la tabla codes
## Mantener los individuos que tengan un Code_comm no nulo
df_codes_clean = df_codes.copy()
df_codes_clean = df_codes_clean[df_codes_clean['Code_comm'].notnull()]
df_codes_clean
```

```python
## Crear una función que permita extraer los códigos de productos
def clean_code(text):
    text = str(text)
    parent_code = None
    if len(text) == 11:
        code = text[:5]
        parent_code = text[:1]
    else:
        code = text[:6]
        parent_code = text[:2]
    try:
        parent = df_parents[df_parents['Code_comm'] == parent_code]['Description'].values[0]
    except:
        parent = None
    return (code, parent)
```

```python
df_codes_clean[['clean_code', 'parent_description']] = df_codes_clean.apply(lambda x: clean_code(x['Code']), axis=1, result_type='expand')
df_codes_clean  
```

```python
df_codes_clean = df_codes_clean[df_codes_clean['clean_code'].notnull()][['clean_code', 'Description', 'parent_description']]
df_codes_clean
```

```python
dim_codes = df_codes_clean.copy()
dim_codes['id_code'] = df_codes_clean.index
dim_codes['clean_code'] = dim_codes['clean_code'].astype('int64')
dim_codes
```

```python
#2. Limpiar la tabla countries
df_countries_clean = df_countries.copy()
df_countries_clean = df_countries_clean[['alpha-3', 'country', 'region', 'sub-region' ]]
df_countries_clean = df_countries_clean[df_countries_clean['alpha-3'].notnull()]
dim_countries = df_countries_clean.copy()
dim_countries['id_country'] = df_countries_clean.index
dim_countries
```

```python
#3 Adicionar campos de dim_codes y dim_countries a df_trades
df_trades_clean = df_trades.merge(dim_codes[['clean_code', 'id_code']], how='left', left_on='comm_code', right_on='clean_code')
df_trades_clean
```

```python
df_trades_clean = df_trades_clean.merge(dim_countries[['alpha-3', 'id_country']], how='left', left_on='country_code', right_on='alpha-3')
```

```python
#4 construir una función que construya las demás dimensiones
def create_dimensiones(data, id_name):
    list_keys = []
    value = 1
    for key in data:
        list_keys.append(value)
        value = value + 1
    
    return pd.DataFrame({id_name:list_keys, 'values':data})
```

```python
#5 crear las dimensiones quantity
dim_quantity = create_dimensiones(df_trades_clean['quantity_name'].unique(), 'id_quantity')
dim_quantity
```

```python
#6 crear la dimensión flow
dim_flow = create_dimensiones(df_trades_clean['flow'].unique(), 'id_flow')
dim_flow
```

```python
#7 crear la dimensión temporal (year)
dim_time = create_dimensiones(df_trades_clean['year'].unique(), 'id_time')
dim_time
```

```python
#8 Mezclar las dimensiones time, quantity y flow con la tabla trades_clean
df_trades_clean = df_trades_clean.merge(dim_quantity, how='left', left_on='quantity_name', right_on='values')
df_trades_clean = df_trades_clean.merge(dim_flow, how='left', left_on='flow', right_on='values')
df_trades_clean = df_trades_clean.merge(dim_time, how='left', left_on='year', right_on='values')
df_trades_clean
```
